# Logistic Regression

```python
X = df.loc[:, df.columns!='Personal Loan']
y = df['Personal Loan']
```

```python
X_encoded = pd.get_dummies(columns = ['Home Ownership'] , data= X, drop_first = True)
X_encoded.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded,y_encoded, test_size=0.2, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, precision_score, recall_score, f1_score, roc_auc_score, classification_report

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap="Blues")
```

```python
print("Precision score=", precision_score(y_test, y_pred))
print("Recall score=",recall_score(y_test, y_pred))
print("f1 score=", f1_score(y_test, y_pred))
print("auc_roc=", roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

# SVC

```python
X = df.loc[:, df.columns!='Personal Loan']
y = df['Personal Loan']

X_encoded = pd.get_dummies(columns = ['Home Ownership'] , data= X, drop_first = True)

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded,y_encoded, test_size=0.2, random_state=40)

from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import ConfusionMatrixDisplay, precision_score, recall_score, f1_score, roc_auc_score,  classification_report

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap="Blues")
```

```python
print("Precision score=", precision_score(y_test, y_pred))
print("Recall score=",recall_score(y_test, y_pred))
print("f1 score=", f1_score(y_test, y_pred))
print("auc_roc=", roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import ConfusionMatrixDisplay, precision_score, recall_score, f1_score, roc_auc_score, classification_report

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap="Blues")
```

```python
print("Precision score=", precision_score(y_test, y_pred))
print("Recall score=",recall_score(y_test, y_pred))
print("f1 score=", f1_score(y_test, y_pred))
print("auc_roc=", roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

# Decision Trees 

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import ConfusionMatrixDisplay, precision_score, recall_score, f1_score, roc_auc_score,  classification_report

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap="Blues")
```

```python
print("Precision score=", precision_score(y_test, y_pred))
print("Recall score=",recall_score(y_test, y_pred))
print("f1 score=", f1_score(y_test, y_pred))
print("auc_roc=", roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

Here, f1 score and AUC-ROC (Area Under the Receiver Operating Characteristic Curve) are highest in Decision Trees Model. So, we choose this model.
